```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
import pandas as pd
import os
```

```python
df = df.sample(frac=1)
```

```python
input_cols = df.columns[5:].tolist()
input = df[input_cols]
target = df["Label"]
```

Standardizing the data


```python
scaler = StandardScaler()
scaler.fit(input)
input = scaler.transform(input)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(input, target, test_size=0.2, random_state=42)
```

Gradient boosting classifier

```python
clf = GradientBoostingClassifier(n_estimators=50, learning_rate=0.1, random_state=42)
clf.fit(X_train, y_train)
y_preds = clf.predict(X_train)
print("Training Accuracy: ", clf.score)
print("Training Classification Accuracy Score: ", accuracy_score(y_train, y_preds))
print("Precision Score: ", precision_score(y_train, y_preds))
print("Recall Score: ", recall_score(y_train, y_preds))
y_valid_preds = clf.predict(X_valid)
print("Validation Classification Accuracy Score: ", accuracy_score(y_valid, y_valid_preds))
print("Precision Score: ", precision_score(y_valid, y_valid_preds))
print("Recall Score: ", recall_score(y_valid, y_valid_preds))
```

LDA

```python
lda = LinearDiscriminantAnalysis(n_components=1)
X_train = lda.fit_transform(X_train, y_train)
X_valid = lda.transform(X_valid)
```

```python
clf1 = GradientBoostingClassifier(n_estimators=30, learning_rate=0.1,random_state=42)
clf1.fit(X_train, y_train)
y_preds = clf1.predict(X_train)
print("Training Classification Accuracy Score: ", accuracy_score(y_train, y_preds))
print("Precision Score: ", precision_score(y_train, y_preds))
print("Recall Score: ", recall_score(y_train, y_preds))
y_valid_preds = clf1.predict(X_valid)
print("Validation Classification Accuracy Score: ", accuracy_score(y_valid, y_valid_preds))
print("Precision Score: ", precision_score(y_valid, y_valid_preds))
print("Recall Score: ", recall_score(y_valid, y_valid_preds))
```

```python
def check_scores(**params):
  clf = GradientBoostingClassifier(**params)
  clf.fit(X_train, y_train)
  y_preds = clf.predict(X_train)
  train_acc = accuracy_score(y_train, y_preds)
  train_prec = precision_score(y_train, y_preds)
  train_rec = recall_score(y_train, y_preds)

  y_valid_preds = clf.predict(X_valid)
  val_acc = accuracy_score(y_valid, y_valid_preds)
  val_prec = precision_score(y_valid, y_valid_preds)
  val_rec = recall_score(y_valid, y_valid_preds)
  return [train_acc, train_prec, train_rec, val_acc, val_prec, val_rec]
```

```python
est_list = [10,20,30,40,50,60,70,80,90]
learning_rate = [0.5,0.3,0.1,0.01]
```

```python
for est in est_list:
  print("n_estimators: ", est)
  scores_array = []
  lr_list = []
  for lr in learning_rate:
    print("Learning Rate: ", lr)
    scores = check_scores(n_estimators=est, learning_rate=lr)
    scores_array.append(scores)
    lr_list.append(lr)
    print("\n")
  df = pd.DataFrame(data=scores_array, columns=["Train Accuracy Score", "Precision Score", "Recall Score", "Val Accuracy Score", "Precision Score", "Recall Score"], index=lr_list)
  df["n_estimators"] = est
  df.to_csv(f"file_{est}.csv", index=True)
  print("\n")
```
